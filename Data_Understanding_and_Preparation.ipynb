# Visualizing the distributions of sales and advertising budgets using histograms

```python
import matplotlib.pyplot as plt
import pandas as pd


# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.hist(df['TV'], bins=10, color='blue', edgecolor='black')
plt.title('TV Advertising Budget Distribution')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(df['Radio'], bins=10, color='green', edgecolor='black')
plt.title('Radio Advertising Budget Distribution')
plt.xlabel('Radio Advertising Budget')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(df['Newspaper'], bins=10, color='orange', edgecolor='black')
plt.title('Newspaper Advertising Budget Distribution')
plt.xlabel('Newspaper Advertising Budget')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(df['Sales'], bins=10, color='red', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Visualizing the distributions of sales and advertising budgets using Boxplot


import seaborn as sns


# Plot box plots
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
sns.boxplot(data=df['TV'], color='blue')
plt.title('TV Advertising Budget Distribution')

plt.subplot(2, 2, 2)
sns.boxplot(data=df['Radio'], color='green')
plt.title('Radio Advertising Budget Distribution')

plt.subplot(2, 2, 3)
sns.boxplot(data=df['Newspaper'], color='orange')
plt.title('Newspaper Advertising Budget Distribution')

plt.subplot(2, 2, 4)
sns.boxplot(data=df['Sales'], color='red')
plt.title('Sales Distribution')

plt.tight_layout()
plt.show()

```
