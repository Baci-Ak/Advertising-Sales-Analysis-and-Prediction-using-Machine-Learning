```python
import matplotlib.pyplot as plt
import warnings
```

```python
#Loading the dataset

import pandas as pd
df = pd.read_csv("Advertising.csv", usecols=lambda column: column != 'Unnamed: 0')
df.head()
```

# Bivariate Analysis

```python
# Visualizing relationships between sales and each advertising medium (TV, radio, newspaper) using scatter plots.



# Create scatter plots
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.scatter(df['TV'], df['Sales'], color='blue', alpha=0.7)
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Relationship: TV vs Sales')

plt.subplot(1, 3, 2)
plt.scatter(df['Radio'], df['Sales'], color='green', alpha=0.7)
plt.xlabel('Radio Advertising Budget')
plt.ylabel('Sales')
plt.title('Relationship: Radio vs Sales')

plt.subplot(1, 3, 3)
plt.scatter(df['Newspaper'], df['Sales'], color='orange', alpha=0.7)
plt.xlabel('Newspaper Advertising Budget')
plt.ylabel('Sales')
plt.title('Relationship: Newspaper vs Sales')

plt.tight_layout()
plt.show()
```

```python
#Calculate and visualize correlation coefficients to understand the strength of relationships.

import seaborn as sns


# Calculate correlation matrix
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

# Multivariate Analysis

```python
#Use pair plots to visualize correlations between all features.


# Create pair plots
sns.pairplot(df, kind='scatter', diag_kind='kde', hue='Sales', palette='coolwarm')
plt.suptitle('Pair Plots of Features', y=1.02)
plt.show()
```

```python
# Explore interactions between different advertising media.

import warnings


# Select the advertising media for exploration
media_to_explore = ['TV', 'Radio', 'Newspaper']

# Create pair plots for media interactions
sns.pairplot(df[media_to_explore], kind='scatter', diag_kind='kde')
plt.suptitle('Interactions between Advertising Media', y=1.02)
plt.show()
```
